```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# rice.png is directly in the question folder
Q_DIR    = os.path.normpath(os.path.join(os.getcwd(), '..', 'question'))
SAVE_DIR = os.path.join(os.getcwd(), 'outputs')
os.makedirs(SAVE_DIR, exist_ok=True)
```

```python
# load as grayscale, then apply both filters with the same 5x5 kernel
# sigma=1.5 for gaussian; median uses the same window size
img    = cv2.imread(os.path.join(Q_DIR, 'rice.png'), cv2.IMREAD_GRAYSCALE)
gauss  = cv2.GaussianBlur(img, (5, 5), 1.5)
median = cv2.medianBlur(img, 5)
```

```python
# side by side so its easy to see how each filter handles the salt & pepper noise
# median should remove the speckles cleanly; gaussian will blur them instead
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, im, t in zip(axes,
                     [img, gauss, median],
                     ['Noisy (salt & pepper)', 'Gaussian  5x5', 'Median  5x5']):
    ax.imshow(im, cmap='gray')
    ax.set_title(t)
    ax.axis('off')

plt.tight_layout()
plt.savefig(os.path.join(SAVE_DIR, 'q8_noise_filter.png'), dpi=150)
plt.show()
```
